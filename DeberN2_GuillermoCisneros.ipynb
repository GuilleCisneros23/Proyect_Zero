## 3. Given a linked list, detect if the list has a cycle. If a cycle is detected, return the position of the node (with respect to the head) where the cycle starts.

For example:

![](./cycle.png)

**Input:**
- Jan -> Feb -> March -> Dec

**Output:**
- True
- 2

```python
class Node:
 
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
 
    def __init__(self):
        self.head = None
        
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
    
    def detectLoop(self):
        s = set()
        temp = self.head
        while (temp):

            if (temp in s):
                return True
            s.add(temp)
            temp = temp.next
 
        return False
```
